```python
import pandas as pd
import numpy as np
```

```python
Amazonsales= pd.read_csv('Amazon Sale Report.csv')
Amazonsales.head()
```

```python
# Drop the last column
Amazonsales.drop(Amazonsales.columns[-1], axis=1, inplace=True)

# Verify the change
print("Updated Columns:")
print(Amazonsales.columns)
```

```python
# Check the data types of each column
print("Data Types of Each Column:")
print(Amazonsales.dtypes)
```

```python
# Correct Data Types

# Convert 'Date' column to datetime
Amazonsales['Date'] = pd.to_datetime(Amazonsales['Date'], errors='coerce')

# Convert 'ship-postal-code' to string (object)
Amazonsales['ship-postal-code'] = Amazonsales['ship-postal-code'].astype(str)

# Verify the corrected data types
print("\nData Types After Correction:")
print(Amazonsales.dtypes)
```

```python
# Check for duplicate rows
duplicate_count = Amazonsales.duplicated().sum()

# Print the number of duplicate rows
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
# Check for null values in each column
null_values = Amazonsales.isnull().sum()

# Print the null values for each column
print("Null Values in Each Column:")
print(null_values)
```

```python


# 1. Fill 'Courier Status' with 'Not Available'
Amazonsales['Courier Status'].fillna('Not Available', inplace=True)

# 2. Fill 'currency' with the mode (most common value)
if 'currency' in Amazonsales.columns:
    currency_mode = Amazonsales['currency'].mode()[0]
    Amazonsales['currency'].fillna(currency_mode, inplace=True)

# 3. Replace nulls in 'Amount' with 0
Amazonsales['Amount'].fillna(0, inplace=True)

# 4. Fill 'ship-city', 'ship-state', 'ship-country' with 'Unknown'
Amazonsales['ship-city'].fillna('Unknown', inplace=True)
Amazonsales['ship-state'].fillna('Unknown', inplace=True)
Amazonsales['ship-country'].fillna('Unknown', inplace=True)

# 5. Replace nulls in 'promotion-ids' with 'No Promotion'
Amazonsales['promotion-ids'].fillna('No Promotion', inplace=True)

# 6. Fill 'fulfilled-by' with 'Not Specified'
Amazonsales['fulfilled-by'].fillna('Not Specified', inplace=True)

# Verify changes
print("Null Values After Cleaning:")
print(Amazonsales.isnull().sum())
```

```python
# Save the cleaned dataset to a new CSV file
cleaned_file_path = 'cleaned_amazon_sale_report.csv'  # Replace with your desired file name
Amazonsales.to_csv(cleaned_file_path, index=False)

print(f"Cleaned dataset saved as '{cleaned_file_path}'")
```
